```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv('/content/wine-clustering.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df1=df.drop(columns=['Magnesium','Proline'])
```

```python
df1.head()
```

```python
mcss=[]

for i in range(1,11):
  kmeans=KMeans(n_clusters=i)
  kmeans.fit(df1)
  mcss.append(kmeans.inertia_)
```

```python
mcss
```

```python
range(1,11)
```

```python
plt.plot(range(1,11),mcss)
```

```python
model=KMeans(n_clusters=4)
model.fit(df1)
```

```python
x=model.labels_
```

```python
plt.scatter(x,df1['Alcohol'])
plt.scatter(x,df1['Malic_Acid'],color='red')
plt.scatter(x,df1['Ash_Alcanity'],color='yellow')
```
