```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.model_selection import GridSearchCV
```

```python
df=pd.read_csv('/content/Car_Price_Prediction.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder , StandardScaler
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop('Price',axis=1),df['Price'],test_size=0.3,random_state=42)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
x_train.head()
```

```python
trf=ColumnTransformer(
    transformers=[
        ('ohe',OneHotEncoder(handle_unknown='ignore'),[0,1,5,6]),
        ('scaler',StandardScaler(),[2,4])
    ]
,remainder='passthrough')
```

```python
pipe=Pipeline(
    steps=[
        ('trf',trf),
        ('dt',DecisionTreeRegressor(max_depth=7))
    ]
)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
model=BaggingRegressor(estimator=DecisionTreeRegressor(max_depth=7),
                       n_estimators=20
                       )
```

```python
pipe1=Pipeline(steps=[
    ('trf',trf),
    ('model',model)
])
```

```python
pipe1.fit(x_train,y_train)
```

```python
y_pred1=pipe1.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

**Score of Using DecisionTree**

```python
r2_score(y_test,y_pred)
```

**Score of Using Bagging ensemble**

```python
r2_score(y_test,y_pred1)
```
